# 1.3 Data Preparation

```python
# 1.3.1  Testing Set
# Storing all the rows and columns apart fromm the Approved Column in X
X = df.iloc[:, :-1]

# Storing the Approved Column in
y = df.iloc[:, -1]
```

```python
# In order to use the categorical labels, we need to encode them into integers
# I would ideally not do this for RandomForestClassifier but for the purposes of good practice I will
# Checking to see if the Approval values were replace from before hand
y_labels = y

print(y_labels.head())
```

```python
# encoding the categorical variables in X into integers
from sklearn.preprocessing import LabelEncoder

X_labels = X
le = LabelEncoder()

for column in X.columns:
    if X_labels[column].dtype == 'object':
        X_labels[column] = le.fit_transform(X[column])



print(X_labels.head())

# Below is a sample output after encoding the categorical variables in X into integers
```

```python
# Checking if all the inputs are all Integers or float
X_labels.info()
```

Below shows,
- Splitting the data into training and testing sets

- Using the tain_test_split

```python
# I'm using the X_labels and y_labels because it has the updated data

from sklearn.model_selection import train_test_split
X_labels_train, X_labels_test, y_labels_train, y_labels_test = train_test_split(X_labels, y_labels, test_size = 0.3, random_state = 30)
#type(X_labels_train)

# The train_test_split function above splits them into X_labels_train, X_labels_test, y_labels_train, y_labels_test
```

```python
# Checking if the y_labels_train data split is well represented in the training
y_labels_train.value_counts(normalize = True)
```

From the above results, the y_labels_train data split is well represented

```python
# Checking if the y_labels_test data split is well represented in the testing
y_labels_test.value_counts(normalize = True)
```

From the above results, the y_labels_test data split is well represented

```python
# Feature Scaling
# The Min-Max Approach
X = X_labels

# Just giving the values in X_labels to a variable X
# This is make it simple to use
X.head() # Display the first few columns in the new variable X
```

```python
from pandas.core.tools.datetimes import Scalar
# Using the Scikit-Learn to perfrom features standardisation

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# transform the data
scaled = scaler.fit_transform(X_labels_train)

# Store the results into a dataframe called X_trains

X_trains = pd.DataFrame(scaled, columns = X_labels_train.columns)

# scale the testing set
scaler = StandardScaler()

# tranform the data
scaled = scaler.fit_transform(X_labels_test)

# store the data into datafrane called X_tests
X_tests = pd.DataFrame(scaled, columns = X_labels_test.columns)
```

```python
# Checking the new columns for X_trains
X_trains.describe()
```

```python
# Checking the new columns for X_tests
X_tests.describe()
```

# 2.0 Modelling / classification
2.1 Building the model using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# create an instance of RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100,)

# fitting the model on the training data
rf.fit(X_trains, y_labels_train)
```

Making the predictions

```python
# Predicted classes of the tessting set (X)
rf_predictions = rf.predict(X_tests)

# getting the probability of the predictions
rf_probs = rf.predict_proba(X_tests)

# Let's predict the class probabilities for the first observation in the testing set.
# This is to see the model's confidence in its predictions for a specific data point.
rf_probs[:1]

```

# 2.2 Results

```python
# Computing the accuracy of the model

import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

accuracy = accuracy_score(y_labels_test, rf_predictions)

# A print statement to see the accuracy of the model under 2 decimal points

print(f'The overall accuracy of RF is {np.round(accuracy * 100, 2)}%')
```

The accuracy of the model above is 88.78% which is quite good.

```python
# The confusion metrics
# Printing the confusion matrix of the Approval Card

print(confusion_matrix(y_labels_test, rf_predictions))
```

Since the confusion metrix summarizes the model's performance on the test set.

From the results above:

- True Negative (TN): 93 cases were correctly predicted as the negative class.
- False Positive (FP): 12 cases were incorrectly predicted as the positive class when they were actually the negative class.
- False Negative (FN): 10 cases were incorrectly predicted as the negative class when they were actually the positive class.
- True Positive (TP): 81 cases were correctly predicted as the positive class.

```python
# using the classification report to show the main classification metrics per class label
# classification_report

print(classification_report(y_labels_test, rf_predictions))
```

In summary, Random Forest classifier is performing well with an overall accuracy of 89%. Precision, recall, and F1-score are high for both classes, indicating a balanced and effective model. These metrics provide a comprehensive view of the model's performance on each class and across all classes.
